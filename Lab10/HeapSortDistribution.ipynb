```
#Tomado de geekforgeeks
def heapify(arr, N, i):
    largest = i  # Initialize largest as root
    l = 2 * i + 1     # left = 2*i + 1
    r = 2 * i + 2     # right = 2*i + 2

    # See if left child of root exists and is
    # greater than root
    if l < N and arr[largest] < arr[l]:
        largest = l

    # See if right child of root exists and is
    # greater than root
    if r < N and arr[largest] < arr[r]:
        largest = r

    # Change root, if needed
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # swap

        # Heapify the root.
        heapify(arr, N, largest)

# The main function to sort an array of given size


def heapSort(arr):
    N = len(arr)

    # Build a maxheap.
    for i in range(N//2 - 1, -1, -1):
        heapify(arr, N, i)

    # One by one extract elements
    for i in range(N-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  # swap
        heapify(arr, i, 0)
```

```
A = [1,2,3,4,5,6,7,8]
heapSort(A)
print(A)

A = [6,7,2,5,4,3,8,1]
heapSort(A)
print(A)

A = [8,7,6,5,4,3,2,1]
heapSort(A)
print(A)
```

```
def heapify(arr, n, i, steps):
    largest = i  # Initialize largest as root
    left = 2 * i + 1
    right = 2 * i + 2

    # Check if left child of root exists and is greater than root
    if left < n and arr[i] < arr[left]:
        largest = left

    # Check if right child of root exists and is greater than the largest so far
    if right < n and arr[largest] < arr[right]:
        largest = right

    # Swap root with the largest element if necessary
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        steps += 1

        # Heapify the affected sub-tree recursively
        steps = heapify(arr, n, largest, steps)

    return steps

def heapSort_steps(arr):
    n = len(arr)
    steps = 0

    # Build a max-heap
    for i in range(n // 2 - 1, -1, -1):
        steps = heapify(arr, n, i, steps)

    # Extract elements from the heap one by one
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        steps += 1

        # Heapify the reduced heap
        steps = heapify(arr, i, 0, steps)

    return steps
```

# Insertion sort  time complexity histograms for permutations of $n$ elements with repetitions counting steps like CLRS I.2.1

$$t (n) =  2n - 1  + 3 \sum_{j=2,..,n} t_j $$

## Best case - increasingly sorted array
$$t_b(n) =  5n - 4$$

## Worst case - de creasingly sorted array

## Average case - expected number of comparisons + swaps on random unifrom permutation of $n$ different elements

```
# Average case on random uniform permtations with repetitions of  $n$   elements  $t_a(n) = \frac{3}{4}n^2 + \frac{7}{2}n -\frac{13}{4}= \Theta(n^2) $
import numpy as np
import pandas as pd
import scipy.linalg as la

n = np.array([3.0,4.0,5.0,6.0])    # Number n of sorted elments
t = np.array([14.0,22.75,33.0,44.75]) # Average comparisons + swaps for the corresponding n

data = pd.DataFrame({'n': n, 't': t})
data
```

```
X = np.column_stack([n**0,n**1,n**2,n**3])
a = la.solve(X,t)
a
```
